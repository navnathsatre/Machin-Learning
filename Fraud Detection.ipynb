### Input Split

```python
X = df.drop(columns=['Class'], axis=1)
y = df['Class']
```

### Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaler = sc.fit_transform(X)
```

```python
x_scaler[-1]
```

### Model Training

```python
# train test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score, confusion_matrix
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y, test_size=0.25, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# training
model.fit(x_train, y_train)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_jobs = -1)
# training
model.fit(x_train, y_train)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

### Class Imbalancement

```python
sns.countplot(y_train)
```

```python
pd.value_counts(y_train)
```

```python
# hint - use combination of over sampling and under sampling
# balance the class with equal distribution
from imblearn.over_sampling import SMOTE
over_sample = SMOTE()
x_smote, y_smote = over_sample.fit_resample(x_train, y_train)
```

```python
sns.countplot(y_smote)
```

```python
pd.value_counts(y_smote)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_jobs=-1)
# training
model.fit(x_smote, y_smote)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV
params = [{"n_estimators":[150, 200, 300]},{"criterion":['gini','entropy']},{"max_depth":[8,7,6,5]},{"max_features" : ["auto", "sqrt", "log2"]}]
search = RandomizedSearchCV(model, params,scoring='accuracy')
```
