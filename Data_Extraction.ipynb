```python
#from textblob import TextBlob
import tweepy
#import sys
import pandas as pd
```

```python
mykeys = open('tweetfile.txt','r').read().splitlines()
api_key = mykeys[0]
api_key_secret = mykeys[1]
access_token = mykeys[2]
access_token_secret = mykeys[3]
```

```python
auth = tweepy.OAuthHandler(consumer_key = api_key, consumer_secret = api_key_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
lst = []
search_term = 'trending news'
tweet_amount = 100
for i in tweepy.Cursor(api.search, q=search_term, lang='en').items(tweet_amount):
    lst.append(i.text)
    #print(i.text)
df = pd.DataFrame({'tweet':lst})
df
```
